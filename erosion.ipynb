```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Cargar la imagen
img = cv2.imread('/content/morelos.jpg', 0)

# Crear un elemento estructurante con forma de cruz y tama√±o 3x3
kernel = np.ones((3,3),np.uint8)

# Aplicar erosi√≥n a la imagen
erosion = cv2.erode(img, kernel, iterations = 1)

# Mostrar la imagen resultante
comparacion = cv2.hconcat([img, erosion])
print("---------------------------------------Original--------------------------------|--------------------------------------Erosi√≥n----------------------------------")
cv2_imshow(comparacion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
